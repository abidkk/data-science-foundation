# Python List Methods


## 1. Append : Add an element at the end of the list:
- takes only one arguement (value, list, tuple)

```python
nums = [30,40,10,60,20,80,50,70,90,50,20]
print(nums)
```

```python
nums.append(100)
print(nums)
```

```python
nums.append([200,300,400])
print(nums)
```

```python
nums.append((111,222))
print(nums)
```

## 2. Clear: Removes all element from the list
- returns empty list

```python
nums.clear()
print(nums)
```

## 3. Copy: returns a copy of a list

### 🔹 Shallow Copy 
A shallow copy makes a new box,
but the items inside are the same objects.

If the items are simple numbers → no issue.
If the items are lists (nested) → both share them.


### 🔹 Deep Copy (Simple Meaning)

A deep copy makes a new box
and also makes new copies of everything inside it.

Nothing is shared.


#### 🔥 The Difference in One Line

Shallow copy → copies the container only

Deep copy → copies container + everything inside

### Examples of Shallow Copies

```python
cities = ["Karachi", "Lahore", "Hyderabad", "Sialkot"]
c1 = cities.copy()
print(c1)
```

```python
c1.append("Islamabad")
print(c1)
print(cities)
```

```python
cities[0]="0000"
```

```python
print(c1)
```

```python
l1 = [100,200 ,300 , [400,500]]
print(l1)
```

```python
l2 = l1.copy()
l2.append(200100)
print(l2)
```

```python
print(l1)
```

```python
l2[3][1]=23112
```

```python
print(l1)
```

## 4. count()	
Returns the number of elements with the specified value
takes one argument

```python
list1 = [200,300,400,500,200,400,700]
print(list1.count(3)) # not found returns 0
print(list1.count(400)) # found twice
```

## 5. extend()	
Add the elements of a list (or any iterable), to the end of the current list

```python
l1 = [200,300,400,500]
# l1.extend(1000) ## type error ( accept iterable values)

l1.extend([600,800])
print(l1)
```

## 6. index()	
Returns the index of the first element with the specified value

```python
nums = [500,100,400,200,500,300,500,400]
print(nums)
```

```python
print(nums.index(500)) # first 500 found on 0th index
print(nums.index(400)) # first 400 found on 2nd index
# print(nums.index(900)) # if not found value error
```

## 7. insert()	
Adds an element at the specified position
- takes two arguments (INDEX VALUE)

```python
l1 = [30,40,10,20,40,50]
print(l1)

# insert
l1.insert(1, 300) ## insert 300 on 1st index
l1
```

```python
# insert
l1.insert(4, [111,222]) ## append list on 4th index
l1
```

## 8. pop()	
Removes the element at the specified position

```python
l1 = [20,30,40,50,60,70]
l1.pop() ## 70 deleted
print(l1)
```

## 9. remove()	
Removes the first item with the specified value

```python
print(l1)
# l1.remove(10) # value error if value not found

l1.remove(40) # 40 removed
print(l1)
```

## 10. reverse()	
Reverses the order of the list

```python
nums = [20,40,60,70,10,30,50]
print(nums)
nums.reverse()
print(nums)
```

```python
colors = ["white", "blue", "orange", "purple"]
print(colors)
colors.reverse()
print(colors)
```

## 11. sort()	
Sorts the list

```python
nums = [20,40,60,70,10,30,50]
print(nums)
nums.sort()
print(nums)

## sort in descending
nums.sort(reverse=True)
print(nums)
```

```python
colors = ["white", "blue", "orange", "purple"]
print(colors)
colors.sort()
print(colors)
## sort in descending
colors.sort(reverse=True)
print(colors)
```
